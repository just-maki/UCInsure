```python
# California Wildfire Datasets - Data Cleaning and Merging

#DataSets:
#DataSet1 - CAL Fire Perimeters
#DataSet2 - CAL Fire DINS
#DataSet3 - CAL Fire Damage
```

```python
# Import all required libraries

import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded successfully")
print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
```

```python
# Set paths to your CSV files

BASE_PATH = "../data/raw/"
OUTPUT_PATH = "../data/processed/"

# Create processed folder if it doesn't exist
os.makedirs(OUTPUT_PATH, exist_ok=True)

# File paths
PERIMETERS_FILE  = BASE_PATH + "calfire_perimeters.csv"
DINS_FILE        = BASE_PATH + "calfire_dins.csv"
DAMAGE_FILE      = BASE_PATH + "ca_wildfire_damage.csv"

print("Paths set successfully")
print(f"Looking for files in: {BASE_PATH}")
```

```python
# Load all 3 CSVs

df1 = pd.read_csv(PERIMETERS_FILE)
df2 = pd.read_csv(DINS_FILE)
df3 = pd.read_csv(DAMAGE_FILE)

print("All 3 datasets loaded!")
print(f"\nDataset 1 - Fire Perimeters : {df1.shape[0]:,} rows, {df1.shape[1]} columns")
print(f"Dataset 2 - DINS            : {df2.shape[0]:,} rows, {df2.shape[1]} columns")
print(f"Dataset 3 - Wildfire Damage : {df3.shape[0]:,} rows, {df3.shape[1]} columns")
```

```python
# Quick look at each dataset
print("=" * 60)
print("DATASET 1 — CAL FIRE PERIMETERS (first 3 rows)")
print("=" * 60)
display(df1.head(3))

print("=" * 60)
print("DATASET 2 — DINS STRUCTURE DAMAGE (first 3 rows)")
print("=" * 60)
display(df2.head(3))

print("=" * 60)
print("DATASET 3 — CA WILDFIRE DAMAGE (first 3 rows)")
print("=" * 60)
display(df3.head(3))
```

```python
# Check missing values in each dataset
print("=" * 60)
print("MISSING VALUES SUMMARY")
print("=" * 60)

for name, df in [("Dataset 1 - Perimeters", df1),
                 ("Dataset 2 - DINS",       df2),
                 ("Dataset 3 - Damage",     df3)]:
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    pct = (missing / len(df) * 100).round(1)
    print(f"\n{name}:")
    if len(missing) == 0:
        print("  No missing values")
    else:
        for col in missing.index:
            print(f"  {col}: {missing[col]:,} missing ({pct[col]}%)")
```

```python
# Cleaning Each Dataset
```

```python
print("D1 actual columns:", df1.columns.tolist())
```

```python
# Select and rename columns
df1_clean = df1[[
    'Fire Name', 'Year', 'Alarm Date',
    'Containment Date', 'GIS Calculated Acres', 'Cause', 'Agency', 'Unit ID'
]].copy()

df1_clean.columns = [
    'fire_name', 'fire_year', 'alarm_date',
    'containment_date', 'gis_acres', 'cause', 'agency', 'unit_id'
]

# Standardize fire name
df1_clean['fire_name'] = df1_clean['fire_name'].str.upper().str.strip()

# Convert dates
df1_clean['alarm_date']       = pd.to_datetime(df1_clean['alarm_date'], errors='coerce')
df1_clean['containment_date'] = pd.to_datetime(df1_clean['containment_date'], errors='coerce')

# Calculate fire duration (55% missing containment = fill with 0)
df1_clean['fire_duration_days'] = (
    df1_clean['containment_date'] - df1_clean['alarm_date']
).dt.days.fillna(0)

# Fill remaining missing values
df1_clean['gis_acres'] = df1_clean['gis_acres'].fillna(0)
df1_clean['cause']     = df1_clean['cause'].fillna('Unknown')
df1_clean['agency']    = df1_clean['agency'].fillna('Unknown')
df1_clean['fire_name'] = df1_clean['fire_name'].fillna('UNKNOWN')
df1_clean['fire_year'] = df1_clean['fire_year'].fillna(0)

# Derive county from Unit ID
unit_to_county = {
    'AEU':'Amador', 'BDU':'San Bernardino', 'BTU':'Butte',
    'CZU':'Santa Cruz', 'FKU':'Fresno', 'HUU':'Humboldt',
    'KRN':'Kern', 'LAC':'Los Angeles', 'LMU':'Lassen',
    'LNU':'Lake', 'MEU':'Mendocino', 'MRN':'Marin',
    'MVU':'San Diego', 'NEU':'Nevada', 'RRU':'Riverside',
    'SCU':'Santa Clara', 'SHU':'Shasta', 'SKU':'Siskiyou',
    'SLU':'San Luis Obispo', 'SQF':'Tulare', 'TCU':'Tuolumne',
    'TGU':'Tehama', 'VNC':'Ventura', 'ANG':'Los Angeles',
    'THW':'Trinity', 'SIS':'Siskiyou', 'SBC':'Santa Barbara'
}
df1_clean['county'] = df1_clean['unit_id'].map(unit_to_county)
df1_clean['county'] = df1_clean['county'].fillna('Unknown').str.upper().str.strip()

print(f" Dataset 1 cleaned: {df1_clean.shape[0]:,} rows, {df1_clean.shape[1]} columns")
display(df1_clean.head(3))
```

```python
print("D2 actual columns:", df2.columns.tolist())
```

```python
# Select and rename columns
df2_clean = df2[[
    '* Incident Name', 'Incident Start Date', 'County',
    '* Damage', 'Structure Category', '* Structure Type',
    '* Roof Construction', '* Eaves', '* Vent Screen',
    '* Exterior Siding', '* Window Pane',
    'Year Built (parcel)', 'Assessed Improved Value (parcel)',
    'Latitude', 'Longitude'
]].copy()

df2_clean.columns = [
    'fire_name', 'incident_date', 'county',
    'damage_level', 'structure_category', 'structure_type',
    'roof_construction', 'eaves', 'vent_screen',
    'exterior_siding', 'window_pane',
    'year_built', 'assessed_value',
    'latitude', 'longitude'
]

# Standardize fire name and county
df2_clean['fire_name'] = df2_clean['fire_name'].str.upper().str.strip()
df2_clean['county']    = df2_clean['county'].str.upper().str.strip()

# Fix lightning complex naming BEFORE merge (recovers ~5 fires)
df2_clean['fire_name'] = df2_clean['fire_name'].str.replace('CMPLX', 'COMPLEX')
df2_clean['fire_name'] = df2_clean['fire_name'].str.replace('LIGHTNINGCOMPLEX', 'LIGHTNING COMPLEX')

# Extract fire year from incident date
df2_clean['fire_year'] = pd.to_datetime(
    df2_clean['incident_date'], errors='coerce').dt.year

# Map damage level → risk level (TARGET VARIABLE)
damage_map = {
    'DESTROYED (>50%)'  : 'Very High',
    'MAJOR (25-50%)'    : 'High',
    'MINOR (10-25%)'    : 'Medium',
    'AFFECTED (>0-10%)' : 'Low',
    'NO DAMAGE'         : 'Low',
    'INACCESSIBLE'      : 'Low'
}
df2_clean['damage_level'] = df2_clean['damage_level'].str.upper().str.strip()
df2_clean['risk_level']   = df2_clean['damage_level'].map(damage_map)

# Fill missing values
# Text columns → Unknown
for col in ['structure_category','structure_type','roof_construction',
            'eaves','vent_screen','exterior_siding','window_pane']:
    df2_clean[col] = df2_clean[col].fillna('Unknown')

# Numeric columns → median
df2_clean['year_built']     = df2_clean['year_built'].fillna(df2_clean['year_built'].median())
df2_clean['assessed_value'] = df2_clean['assessed_value'].fillna(df2_clean['assessed_value'].median())

print(f" Dataset 2 cleaned: {df2_clean.shape[0]:,} rows, {df2_clean.shape[1]} columns")
print(f"\nRisk Level Distribution:")
print(df2_clean['risk_level'].value_counts())
display(df2_clean.head(3))
```

```python
print("D3 actual columns:", df3.columns.tolist())
```

```python
# Select and rename (adjust column names if different)
df3_clean = df3[[
    'Location', 'Date',
    'Area_Burned (Acres)', 'Homes_Destroyed',
    'Estimated_Financial_Loss (Million $)', 'Fatalities', 'Cause'
]].copy()

df3_clean.columns = [
    'location', 'date',
    'area_burned', 'homes_destroyed',
    'financial_loss_million', 'fatalities', 'cause_d3'
]

# Extract county from Location column
# D3 has no fire_name — county is our only join key
df3_clean['county'] = df3_clean['location'].str.split(',').str[-1]\
                      .str.replace('County','', case=False)\
                      .str.upper().str.strip()

# Extract fire year
df3_clean['fire_year'] = pd.to_datetime(df3_clean['date'], errors='coerce').dt.year

# Fill missing (D3 had no missing values but just in case)
df3_clean['area_burned']            = df3_clean['area_burned'].fillna(0)
df3_clean['homes_destroyed']        = df3_clean['homes_destroyed'].fillna(0)
df3_clean['financial_loss_million'] = df3_clean['financial_loss_million'].fillna(0)
df3_clean['fatalities']             = df3_clean['fatalities'].fillna(0)
df3_clean['cause_d3']               = df3_clean['cause_d3'].fillna('Unknown')

print(f" Dataset 3 cleaned: {df3_clean.shape[0]:,} rows, {df3_clean.shape[1]} columns")
display(df3_clean.head(3))
```

```python
#Merging All 3 Datasets
```

```python
# D2 + D1 on fire_name and fire_year 
merged = pd.merge(
    df2_clean,
    df1_clean[['fire_name', 'fire_year', 'gis_acres',
               'cause', 'agency', 'fire_duration_days']],
    on=['fire_name', 'fire_year'],
    how='left'
)
print(f"After D2 + D1 merge: {merged.shape[0]:,} rows")

# Checking match rate
matched = merged['gis_acres'].notna().sum()
print(f"Rows matched with D1 data: {matched:,}/{len(merged):,} ({matched/len(merged)*100:.1f}%)")

#  D2 + D3 on county + fire_year 
# D3 has no fire_name so we use county as join key

merged = pd.merge(
    merged,
    df3_clean[['county', 'fire_year', 'area_burned',
               'homes_destroyed', 'financial_loss_million', 'fatalities']],
    on=['county', 'fire_year'],
    how='left'
)
print(f"After + D3 merge:     {merged.shape[0]:,} rows")
print(f"Total columns:        {merged.shape[1]}")

# Fill NaNs from unmatched rows
for col in ['gis_acres','fire_duration_days','area_burned',
            'homes_destroyed','financial_loss_million','fatalities']:
    merged[col] = merged[col].fillna(0)

print("\n All 3 datasets merged!")
display(merged.head(3))
```

```python
print("=" * 60)
print("FINAL MERGED DATASET SUMMARY")
print("=" * 60)
print(f"Total rows    : {merged.shape[0]:,}")
print(f"Total columns : {merged.shape[1]}")
print(f"\nColumns:\n{list(merged.columns)}")
print(f"\nRisk Level Distribution:")
print(merged['risk_level'].value_counts())
print(f"\nMissing values remaining:")
remaining = merged.isnull().sum()
print(remaining[remaining > 0] if remaining[remaining > 0].any() else "None")
```

```python
OUTPUT_FILE = OUTPUT_PATH + "merged_calfire.csv"
merged.to_csv(OUTPUT_FILE, index=False)

print(f"Saved to: {OUTPUT_FILE}")
print(f"   Rows    : {merged.shape[0]:,}")
print(f"   Columns : {merged.shape[1]}")
```
